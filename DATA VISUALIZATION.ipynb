```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\\asaha\\Downloads\\titanic.csv")
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Basic information about the dataset
print(df.info())
```

```python
# Describe the dataset to get statistical information
print(df.describe())
```

```python
# Visualizing the number of survivors
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Survived')
plt.title('Number of Survivors')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show() 
```

```python
# Visualizing the distribution of passengers by class
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Pclass')
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing the distribution of passengers by age
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Distribution of Passengers by Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing survival rate by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
# Visualizing survival rate by gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived')
plt.show()
```

```python
# Visualizing the distribution of fares
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', bins=30, kde=True)
plt.title('Distribution of Fares')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the correlation matrix with numeric columns only
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
